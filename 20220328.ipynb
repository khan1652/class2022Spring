```python
def test(rsacGutFsttOgidNm):
  URL = ('http://apis.data.go.kr/1661000/EmergencyStatisticsService/getTrasferPatientByAgeGroupStats?serviceKey=kbZkD1n6DXAZ0QRLp5ENMF2Hy7vrW1kvRQELIbQ8cW06uXNRP%2B2jO6miBBhzwKDVV8T5Om9%2BVv8V6b8HtnRxuQ%3D%3D&numOfRows=100&pageNo=1&resultType=json&sidoHqOgidNm=소방재난본부&rsacGutFsttOgidNm='+rsacGutFsttOgidNm+'&rcptYm&ageScov')

  response = requests.get(URL)
  r = response.json()
  items = r['body']['items']

  ageScov = [item['ageScov'] for item in items]
  rcptYm = [item['rcptYm'] for item in items]
  sidoHqOgidNm = [item['sidoHqOgidNm'] for item in items]
  gutCo = [item['gutCo'] for item in items]

  ageScov = pd.Series(ageScov)
  rcptYm = pd.Series(rcptYm)
  rsacGutFsttOgidNm = pd.Series(rsacGutFsttOgidNm)
  sidoHqOgidNm = pd.Series(sidoHqOgidNm)
  gutCo = pd.Series(gutCo)

  df = pd.DataFrame({ '연령대': ageScov, '접수년월': rcptYm, '시도본부': sidoHqOgidNm, '출동건수': gutCo })
  return df
```

```python
iface = gr.Interface(test, gr.inputs.Dropdown(['동작소방서', '종로소방서']), outputs=['dataframe'])
iface.launch(debug=True)
```
