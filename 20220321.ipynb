```python
import pandas as pd
```

```python
df = pd.read_csv("https://download.mlcc.google.com/mledu-datasets/california_housing_train.csv", sep=",")
df
```

**pd.read_csv**함수는 csv파일을 불러옴. csv는 엑셀 파일 형태(.xls보다 일반적인 형태임). 텍스트가 ","로 정보가 분리됨.
엑셀 형식의 정보를 data frame이라고 함. 

```python
print(df.size); print(df.ndim); print(df.shape)
df.describe()
```

**df.size**는 rows*columns(몇 개의 cell인지 알려줌).
**df.ndim**은 몇 차원인지 알려줌.
**df.shape**는 (rows, columns).
**df.describe()**는 기본적 통계 정보를 알려줌(column별로 평균, 표준편차, 최솟값, 최댓값 등).

```python
df.to_csv('california_housing_train.csv')
```

**df.to_csv**는 변수 형태에서 다시 csv 파일 형태로 저장함

```python
data = {'sepal_length': [6.9, 6.9, 4.8, 5.4, 4.6],
        'sepal_width': [3.2, 3.1, 3.4, 3.0, 3.6],
        'petal_length': [5.7, 5.1, 1.9, 4.5, 1.0],
        'petal_width': [2.3, 2.3, 0.2, 1.5, 0.2],
        'species': ['virginica', 'virginica', 'setosa', 'versicolor', 'setosa']}

df = pd.DataFrame(data)
df
```

data에 dict를 저장한 이후 **pd.DataFrame(data)**를 통해 데이터 프레임을 만들어 줄 수 있음. 위에는 5개의 아이템이 나열돼있고, key(column 이름) : value(list형태)임. value의 수를 각 column마다 동일하게 맞춰줘야 함(위에는 5개).

```python
sepal_length = pd.Series([6.9, 6.9, 4.8, 5.4, 4.6])
sepal_width = pd.Series([3.2, 3.1, 3.4, 3.0, 3.6])
petal_length = pd.Series([5.7, 5.1, 1.9, 4.5, 1.0])
petal_width = pd.Series([2.3, 2.3, 0.2, 1.5, 0.2])
species = pd.Series(['virginica', 'virginica', 'setosa', 'versicolor', 'setosa'])

df = pd.DataFrame({'sepal_length': sepal_length, 'sepal_width': sepal_width, 'petal_length': petal_length, 'petal_width': petal_width, 'SPECIES': species})
df
```

**pd.Series**를 통해 내용을 먼저 구성한 후 **pd.DataFrame()**에 dict를 넣어주면 같은 결과가 출력됨.

```python
df.columns = ['sepal_length', 'sepal_width', 'petal_length', 'petal_width', 'species']
df
```

**df.columns**는 column 명칭을 수정할 수 있음.

```python
df.loc[:, 'sepal_length']  #모든 row, 특정 column
df['sepal_length']
df.sepal_length
```

하나의 column을 가져오는 방법 3가지

```python
df.loc[:, ['sepal_length', 'sepal_width']]
df[['sepal_length', 'sepal_width']]  #list로 묶어줘야 함.
```

두개의 column을 가져올 수 있는 방법 2가지

```python
df.loc[3, :]
```

```python
df.loc[[3, 0], :]
```

```python
df.loc[2:5, :]
```

```python
df[df.sepal_length > 5]  #해당 column 값이 5보다 큰 경우만 가져옴.
```

```python
df.loc[:, 'sepal_length'][4]
df['sepal_length'][4]
df.sepal_length[4]
```

특정 cell을 불러올 수 있는 3가지 방법.

```python
df['new_column'] = 1
df
```

```python
df['new_column'] = [1, 2, 3, 4, 5]
df
```

새로운 column을 더해주는 방법.

```python
df = df.drop('new_column', axis=1) # axis=1: column
df
```

**df.drop**은 column을 삭제하는 방법임. **axis=1**은 'new_column'을 열축에서 찾으라는 의미임.

```python
df.loc[5, :] = ''   #blank string
df
```

새로운 행을 추가하는 방법

```python
df = df.drop(5, axis=0) # axis=0: row
df
```

행을 삭제하는 방법. **axis=0**은 5를 행축에서 찾으라는 의미임.
